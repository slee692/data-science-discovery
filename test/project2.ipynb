# Drug Usage

let's load the dataset showing different types of drug use:

```python
import pandas as pd
df_druguse = pd.read_csv("drug-use-by-age.csv")
df_druguse
```

## Correlation (r)

Let's take a look at the correlations to see if there's any interesting patterns between drug usage!

```python
cor = df_druguse.corr()
cor
```

# Marijuana the Gateway Drug
An interesting pattern is that **Marijuana-use** has a high correlation with several other drugs:
- with hallucinogen-use, r = 0.959353
- with cocaine-use, r = 0.883789	
- with meth-use, r = 0.870446

Let's try plotting the scatter to see the correlation visually:

```python
df_druguse.plot.scatter(y="hallucinogen-use", x="marijuana-use")
```

```python
df_druguse.plot.scatter(y="cocaine-use", x="marijuana-use")
```

```python
df_druguse.plot.scatter(y="meth-use", x="marijuana-use")
```

# Linear Regression
Since each drug has a high correlation (r), let's try creating linear regression models:

## Marijuana-Hallucinogen

```python
from sklearn.linear_model import LinearRegression
model_hal = LinearRegression()
model_hal.fit(df_druguse[["marijuana-use"]], df_druguse["hallucinogen-use"])

b0_hal = model_hal.intercept_
b1_hal = model_hal.coef_
print(b0_hal)
print(b1_hal)
```

Our linear regression model: hallucinogen-use = 0.224*(marijuana-use) - 0.845

This equation will allow us to take the percentage of marijuana use and predict hallucinogen use.

```python
import seaborn as sns
sns.regplot(x=df_druguse["marijuana-use"], y=df_druguse["hallucinogen-use"])
```

## Marijuana-Cocaine

```python
from sklearn.linear_model import LinearRegression
model_coc = LinearRegression()
model_coc.fit(df_druguse[["marijuana-use"]], df_druguse["cocaine-use"])

b0_coc = model_coc.intercept_
b1_coc = model_coc.coef_
print(b0_coc)
print(b1_coc)
```

Our linear regression model: cocaine-use = 0.134*(marijuana-use) - 0.364

This equation will allow us to take the percentage of marijuana use and predict cocaine use.

```python
import seaborn as sns
sns.regplot(x=df_druguse["marijuana-use"], y=df_druguse["cocaine-use"])
```

## Marijuana-Methamphetamine

```python
from sklearn.linear_model import LinearRegression
model_meth = LinearRegression()
model_meth.fit(df_druguse[["marijuana-use"]], df_druguse["meth-use"])

b0_meth = model_meth.intercept_
b1_meth = model_meth.coef_
print(b0_meth)
print(b1_meth)
```

Our linear regression model: meth-use = 0.019*(marijuana-use) + 0.02

This equation will allow us to take the percentage of marijuana use and predict meth use.

```python
import seaborn as sns
sns.regplot(x=df_druguse["marijuana-use"], y=df_druguse["meth-use"])
```

## Conclusion
There appears to be a strong positive relationship between Marijuana-Usage and several other drugs. Although a causal relationship cannot be established, there is an indication that marijuana-use is associated with using other drugs. It would be better to avoid marijuana if one does not want to start other dangerous drugs.
