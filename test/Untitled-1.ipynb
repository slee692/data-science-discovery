```python
#Average GPA per subject
import pandas as pd
df = pd.read_csv("gpa (1).csv")
df = df.groupby(["Subject", "Number"]).agg("sum").reset_index()
df["total"] = df["A+"]*4.0 + df["A"]*4.0 + df["A-"]*3.7 + df["B+"]*3.3 + df["B"]*3.0 + df["B-"]*2.7 + df["C+"]*2.3 + df["C"]*2.0 + df["C-"]*1.7 + df["D+"]*1.3 + df["D"]*1.0 + df["F"]*0
df["number"] = df["A+"] + df["A"] + df["A-"] + df["B+"] + df["B"] + df["B-"] + df["C+"] + df["C"]+ df["C-"]+ df["D+"]+ df["D"] + df["D-"] + df["F"]
df["Average GPA"] = df["total"] / df["number"]
df
```

```python
df = pd.read_csv("gpa.csv")
df_subject = df.groupby("Subject").agg("sum").reset_index()
df_subject["sum GPA"] = df_subject["A+"]*4.00 + df_subject["A"]*4.00 + df_subject["A-"]*3.67 + df_subject["B+"]*3.33 + df_subject["B"]*3.00 + df_subject["B-"]*2.67 + df_subject["C+"]*2.33 + df_subject["C"]*2.00 + df_subject["C-"]*1.67 + df_subject["D+"]*1.33 + df_subject["D"]*1.00 + df_subject["D-"]*0.67 + df_subject["F"]*0.00
df_subject["total num"] = df_subject["A+"] + df_subject["A"] + df_subject["A-"] + df_subject["B+"] + df_subject["B"] + df_subject["B-"] + df_subject["C+"] + df_subject["C"] + df_subject["C-"] + df_subject["D+"] + df_subject["D"] + df_subject["D-"] + df_subject["F"]
df_subject["Average GPA"] = df_subject["sum GPA"] / df_subject["total num"]

df_subject
```
