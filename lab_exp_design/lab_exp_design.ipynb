### 3.2. Puzzle: Creating `block1` and `block2`

This variable is called: Height / ft + in

In our dataset, we have two blocks (or groups) that we can name `block1` and `block2`. Create DataFrames `block1` and `block2` that contain data based on our blocking condition. Remember for this lab, we will use **Height / ft + in** as the blocking condition. Use Python to create `block1` that contains the students who have a *Height / ft + in* greater than 65.
